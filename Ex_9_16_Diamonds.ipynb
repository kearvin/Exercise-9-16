```python
import pandas as pd
import matplotlib.pyplot as plt
```

## a) Load the diamonds.csv dataset

The dataset is loaded directly from a GitHub *raw* CSV URL.
The first column is used as the DataFrame index.


```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv"
df = pd.read_csv(url, index_col=0)

df.head()
```

## b) Display the first seven rows of the DataFrame


```python
df.head(7)
```

## c) Display the last seven rows of the DataFrame


```python
df.tail(7)
```

## d) Descriptive statistics for numerical columns


```python
df.describe()
```

## e) Descriptive statistics for categorical columns


```python
df["cut"].describe()
```

```python
df["color"].describe()
```

```python
df["clarity"].describe()
```

## f) Unique category values


```python
print("Cut:", df["cut"].unique())
print("Color:", df["color"].unique())
print("Clarity:", df["clarity"].unique())
```

## g) Histograms of numerical columns


```python
%matplotlib inline
df.hist(figsize=(10, 8))
plt.show()
```

In order to experience loading and investigating real-world data, Pandas was used to evaluate the diamonds.csv dataset. While descriptive statistics for categorical variables assisted in identifying common categories and distributions, summary statistics produced by the describe() method offered helpful insights into the numerical aspects like price and carat. Many of the variables are skewed rather than evenly distributed, as seen by the histograms, which made it simpler to see how the numerical values are distributed. Overall, this exercise showed how Pandas can effectively use a small number of built-in functions to summarize, analyze, and visualize datasets.
